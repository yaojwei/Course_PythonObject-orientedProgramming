```python
# 
import matplotlib.pyplot as plt
# 
import jieba
# 
from wordcloud import WordCloud

# 
def stopwordslist():
    stopwords = [line.strip() for line in open('ChinsesStopTxt.txt',encoding='UTF-8').readlines()]
    return stopwords

# txt
text = open(r'MergeText.txt',"rb").read()
cut_text= jieba.cut(text.strip())
txt_depart = "/".join(cut_text)
#print(txt_depart)

# 
stopwords = stopwordslist()
ResultTxt = ''
for word in txt_depart.split('/'):
    if word not in stopwords:
        ResultTxt += word
        ResultTxt += "/"

#
bg_pic = imread('man.jpg')

# 
# min_font_size=10)#,mode='RGBA',colormap='pink')
# wc = WordCloud(font_path=r".otf",background_color="white",width=1000,height=860,margin=2,max_font_size=50,max_words=1000)
wc = WordCloud(mask=bg_pic,font_path=r".otf",background_color="white",width=1000,height=860,margin=2)
wc.generate(ResultTxt)
wc.to_file(r"winner.png") 
#
plt.figure("") #
plt.imshow(wc)       # 
plt.axis("off")      #
plt.show()
```
