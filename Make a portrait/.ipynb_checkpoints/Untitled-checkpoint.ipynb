```python
#数学绘图库
import matplotlib.pyplot as plt
#分词库
import jieba
#词云库
from wordcloud import WordCloud
```

```python
# 创建停用词列表
def stopwordslist():
    stopwords = [line.strip() for line in open('ChinsesStopTxt.txt',encoding='UTF-8').readlines()]
    return stopwords
```

```python
#读入txt文本数据
text = open(r'Steve_Jobs_A_Biography.txt',"rb").read()
cut_text= jieba.cut(text.strip())
txt_depart = "/".join(cut_text)
#print(txt_depart)
```

```python
stopwords = stopwordslist()
```

```python
outstr = ''
for word in txt_depart:
    if word not in stopwords:
        outstr += word
        outstr += "/"
```

```python
print(outstr)
```

```python
# 对句子进行中文分词
def seg_depart(text):
    # 对文档中的每一行进行中文分词
    cut_text= jieba.cut(text.strip())
    txt_depart = "/".join(cut_text)
    # 创建一个停用词列表
    stopwords = stopwordslist()
    # 输出结果为outstr
    outstr = ''
   # 去停用词
   for word in txt_depart:
        if word not in stopwords:
            if word != '\t':
                outstr += word
                outstr += " "
    return outstr
```

```python
result = seg_depart(text)
print(result)
```

```python
#生成词云图
# min_font_size=10)#,mode='RGBA',colormap='pink')
# wc = WordCloud(font_path=r"造字工房朗宋（非商用）常规体.otf",background_color="white",width=1000,height=860,margin=2,max_font_size=50,max_words=1000)
wc = WordCloud(font_path=r"造字工房朗宋（非商用）常规体.otf",background_color="white",width=1000,height=860,margin=2)
wc.generate(result)
wc.to_file(r"test.png") 
#显示词云图
plt.figure("词云图") #指定所绘图名称
plt.imshow(wc)       # 以图片的形式显示词云
plt.axis("off")      #关闭图像坐标系
plt.show()
```
