```python
#英文版
# -*- coding: utf-8 -*-
#导入wordcloud模块和matplotlib模块
from wordcloud import WordCloud,ImageColorGenerator
import matplotlib.pyplot as plt
from imageio import imread
#读取一个txt文件
f = open(r'English_Steve_Jobs_A_Biography.txt','r').read()
#读入背景图片
bg_pic = imread('man.jpg')
#生成词云
wordcloud=WordCloud(mask=bg_pic,background_color="white",width=1000,height=860,margin=2 ).generate(f)
#显示词云图片
image_colors = ImageColorGenerator(bg_pic)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
#保存图片
wordcloud.to_file('test.png')
```

```python
#中文版
# -*- coding: utf-8 -*-
#合并一个文件夹下的多个txt文件
import os
#获取目标文件夹的路径
filedir = os.getcwd()+'/中外科学家发明家丛书'
#获取当前文件夹中的文件名称列表
filenames=os.listdir(filedir)
#打开当前目录下的result.txt文件，如果没有则创建
f=open('MergeText.txt','w')
i=0
#先遍历文件名
for filename in filenames:
    i+=1
    print(i)
    if i>0:
        filepath = filedir+'/'+filename
        print(filepath)
        #遍历单个文件，读取行数
        for line in open(filepath,encoding='gbk', errors='ignore'):
            # print(str(line))
            f.writelines(line)
            # f.write('\n')
#关闭文件
f.close()
```

```python
# 数学绘图库
import matplotlib.pyplot as plt
# 分词库
import jieba
# 词云库
from wordcloud import WordCloud
from imageio import imread

# 创建停用词列表
def stopwordslist():
    stopwords = [line.strip() for line in open('ChinsesStopTxt.txt',encoding='UTF-8').readlines()]
    return stopwords

# 读入txt文本数据
text = open(r'MergeText.txt',"rb").read()
cut_text= jieba.cut(text.strip())
txt_depart = "/".join(cut_text)
#print(txt_depart)

# 删除停用词
stopwords = stopwordslist()
ResultTxt = ''
for word in txt_depart.split('/'):
    if word not in stopwords:
        ResultTxt += word
        ResultTxt += "/"

#读入背景图片
bg_pic = imread('man.jpg')

# 生成词云图
# min_font_size=10)#,mode='RGBA',colormap='pink')
# wc = WordCloud(font_path=r"造字工房朗宋（非商用）常规体.otf",background_color="white",width=1000,height=860,margin=2,max_font_size=50,max_words=1000)
wc = WordCloud(mask=bg_pic,font_path=r"造字工房朗宋（非商用）常规体.otf",background_color="white",width=1500,height=1720,margin=2)
wc.generate(ResultTxt)
wc.to_file(r"winner.png") 
#显示词云图
plt.figure("词云图") #指定所绘图名称
plt.imshow(wc)       # 以图片的形式显示词云
plt.axis("off")      #关闭图像坐标系
plt.show()
```
